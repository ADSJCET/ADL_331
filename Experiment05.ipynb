```python
import sklearn.datasets as ds
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.model_selection import train_test_split as tts
dataset = ds.load_breast_cancer()
data, target = dataset.data, dataset.target
x_train, x_test, y_train, y_test = tts(data, target, test_size=0.2, random_state=42)
model = dtc(criterion='entropy', random_state=42)
trained_model = model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print("Accuracy of the model:",round(model.score(x_test,y_test)*100,2),"%\n")
print("-------------------------------------------")
print("Actual\t\t Predicted\t Prediction")
print("-------------------------------------------")
for i in range(len(x_test)):
    print(dataset.target_names[y_test[i]]," \t",dataset.target_names[y_pred[i]],end = "")
    if y_pred[i] == y_test[i]:
        print(" \t","Correct")
    else:
        print(" \t","Incorrect")
```
