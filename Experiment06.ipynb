```python
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
data = pd.read_csv('height-weight.csv')
x=data['height']
y=data['weight']
intercept, slope, r, p, std_err = stats.linregress(x, y)
pred_y = lambda x: x * intercept + slope
test_height = int(input("Enter a sample height: "))
print("Predicted weight: ",round(pred_y(test_height),2))
plt.title("Linear Regression")
plt.xlabel("Height in CM")
plt.ylabel("Weight in KG")
plt.scatter(x,y,color='#5886a5',label = "Data")
plt.plot(x,pred_y(x),color='#004c6d',label = "Regression Line")
plt.scatter(test_height,pred_y(test_height),color='#665191',label="Predicted Value")
plt.legend()
plt.show()
```
